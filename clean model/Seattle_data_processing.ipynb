```python
def fill_propertytype(df):
    for i in range(len(df)):
        if pd.isnull(df['LargestPropertyUseType'][i]):
            df['LargestPropertyUseType'][i] = df['PrimaryPropertyType'][i]
        else:
            continue         

def reclassify_buildingtype(df):
    residential = ['Multifamily LR (1-4)', 'Multifamily MR (5-9)', 'Multifamily HR (10+)']
    specialty = ['Campus', 'SPS-District K-12', 'Nonresidential COS']
    for i in range(len(df)):
        if df['BuildingType'][i] in residential:
            df['BuildingType'][i] = 'Residential'
        elif df['BuildingType'][i] in specialty:
            df['BuildingType'][i] = 'Specialty'
        else:
            continue           

def remove_outliers(df):   
    for i in range(len(df)):
        if pd.isnull(df['Outlier'][i]):
            continue 
        else:
            df.drop([i], inplace=True)     
            

def remove_columns_Seattle_2015(df):
    df.drop(['PrimaryPropertyType',
             'PropertyName',
             'TaxParcelIdentificationNumber', 
             'Location',
             'CouncilDistrictCode',
             'Neighborhood',
             'DefaultData',
             'Comment',
             'ComplianceStatus',
             'Outlier',
             '2010 Census Tracts',
             'Seattle Police Department Micro Community Policing Plan Areas',
             'SPD Beats',
             'NumberofBuildings',
             'NumberofFloors',
             'PropertyGFAParking',
             'PropertyGFABuilding(s)',
             'ListOfAllPropertyUseTypes',
             'LargestPropertyUseTypeGFA',
             'SecondLargestPropertyUseType',
             'SecondLargestPropertyUseTypeGFA',
             'ThirdLargestPropertyUseType',
             'ThirdLargestPropertyUseTypeGFA',
             'YearsENERGYSTARCertified',
             'ENERGYSTARScore',
             'GHGEmissions(MetricTonsCO2e)',
             'GHGEmissionsIntensity(kgCO2e/ft2)',
             'City Council Districts',
             'Electricity(kWh)',
             'NaturalGas(therms)',
             'Zip Codes',
             'OtherFuelUse(kBtu)'
           ], inplace=True, axis=1)            

def remove_columns_Seattle_2016(df):
    df.drop(['PrimaryPropertyType',
             'PropertyName',
             'Address',
             'TaxParcelIdentificationNumber', 
             'CouncilDistrictCode',
             'Neighborhood',
             'Latitude',
             'Longitude',
             'DefaultData',
             'Comments',
             'ComplianceStatus',
             'Outlier',
             'NumberofBuildings',
             'NumberofFloors',
             'PropertyGFAParking',
             'PropertyGFABuilding(s)',
             'ListOfAllPropertyUseTypes',
             'LargestPropertyUseTypeGFA',
             'SecondLargestPropertyUseType',
             'SecondLargestPropertyUseTypeGFA',
             'ThirdLargestPropertyUseType',
             'ThirdLargestPropertyUseTypeGFA',
             'YearsENERGYSTARCertified',
             'ENERGYSTARScore',
             'TotalGHGEmissions',
             'GHGEmissionsIntensity',
             'Electricity(kWh)',
             'NaturalGas(therms)',
           ], inplace=True, axis=1)

def remove_columns_Seattle_2017(df):
    df.drop(['PrimaryPropertyType',
             'PropertyName',
             'Address',
             'TaxParcelIdentificationNumber', 
             'CouncilDistrictCode',
             'Neighborhood',
             'Latitude',
             'Longitude',
             'DefaultData',
             'ComplianceStatus',
             'Outlier',
             'NumberofBuildings',
             'NumberofFloors',
             'PropertyGFAParking',
             'PropertyGFABuilding(s)',
             'ListOfAllPropertyUseTypes',
             'LargestPropertyUseTypeGFA',
             'SecondLargestPropertyUseType',
             'SecondLargestPropertyUseTypeGFA',
             'ThirdLargestPropertyUseType',
             'ThirdLargestPropertyUseTypeGFA',
             'YearsENERGYSTARCertified',
             'ENERGYSTARScore',
             'TotalGHGEmissions',
             'GHGEmissionsIntensity',
             'Electricity(kWh)',
             'NaturalGas(therms)',
           ], inplace=True, axis=1)

def remove_commas(df):
    column_names_1 = np.array(df.columns[7:])
    column_names_1 = np.delete(column_names_1, 1)
    
    if 'BuildingID' in df.columns:
        column_names_1 = np.append(column_names_1, ['BuildingID'])
    
    for i in range(len(column_names_1)):
        column = column_names_1[i]
        df[column] = pd.to_numeric(df[column].astype(str).str.replace(',',''), errors='coerce')
        #df[column] = df[column].replace({r'\\r': ''}, regex=True)
        
def convert_to_int(df):
    if 'BuildingID' in df.columns:
        column_names_2 = ['BuildingID', 'DataYear', 'ZipCode', 'YearBuilt']
    else:
        column_names_2 = ['DataYear', 'ZipCode', 'YearBuilt']
    
    for i in range(len(column_names_2)):
        column = column_names_2[i]
        df[column] = df[column].astype('int64')
        
def change_dtypes(df):
    remove_commas(df)
    convert_to_int(df)
    
def clean_data_Seattle_2015(df): 
    fill_propertytype(df)
    reclassify_buildingtype(df)
    remove_outliers(df)
    remove_columns_Seattle_2015(df)
    df.dropna(inplace=True)
    
def clean_data_Seattle_2016(df): 
    fill_propertytype(df)
    reclassify_buildingtype(df)
    remove_outliers(df)
    remove_columns_Seattle_2016(df)
    df.dropna(inplace=True)
    df.rename(index=str, columns={'OSEBuildingID':'BuildingID'}, inplace=True)
    change_dtypes(df)
    df.set_index('BuildingID', inplace=True)
    
    
def clean_data_Seattle_2017(df): 
    fill_propertytype(df)
    reclassify_buildingtype(df)
    remove_outliers(df)
    remove_columns_Seattle_2017(df)
    df.fillna(0, inplace=True)
    df.rename(index=str, columns={'OSEBuildingID':'BuildingID'}, inplace=True)
    change_dtypes(df)
    df.set_index('BuildingID', inplace=True)
```

```python
clean_data_Seattle_2015(data_2015)

data_2015.set_index('OSEBuildingID', inplace=True)
data_2015 = pd.concat([zipcodes, data_2015], sort=True, axis=1)

data_2015['City'] = 'Seattle'
data_2015['State'] = 'WA'
data_2015 = data_2015[['DataYear',
                       'BuildingType',
                       'City', 
                       'State',
                       'ZipCode',
                       'YearBuilt',
                       'PropertyGFATotal',
                       'LargestPropertyUseType',
                       'SiteEUI(kBtu/sf)',
                       'SiteEUIWN(kBtu/sf)',
                       'SourceEUI(kBtu/sf)',
                       'SourceEUIWN(kBtu/sf)',
                       'SiteEnergyUse(kBtu)',
                       'SiteEnergyUseWN(kBtu)',
                       'SteamUse(kBtu)',
                       'Electricity(kBtu)',
                       'NaturalGas(kBtu)']]

data_2015.dropna(inplace=True)
data_2015.index.rename('BuildingID', inplace=True)
change_dtypes(data_2015)

data_2015.head()
```
